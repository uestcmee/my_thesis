```python
import pandas as pd
import json,re
from bs4 import BeautifulSoup
```

```python
# 总觉得这样不对，但是没找到更好的解析方式
def read_file(file_name='./htmls/0.html'):
    with open (file_name,encoding='utf-8') as f:
        text=f.read()
        lines=(re.findall('(\{"DATADATE".*?"})',text))
        commodity_df=pd.DataFrame(columns=json.loads(lines[0]).keys())

        for i,line in enumerate(lines):
            js=json.loads(line)  
            commodity_df.loc[i]=pd.Series(js)
        return commodity_df
read_file().head()
```

```python
for i in range(26):
    if i ==0:
        com_df=read_file(file_name='./htmls/{}.html'.format(i))
    else:
        com_df=pd.concat([com_df,read_file(file_name='./htmls/{}.html'.format(i))])
com_df
```

```python
com_df.set_index('DATADATE',inplace=True)

```
