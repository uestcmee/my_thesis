```python
import os
from bs4 import BeautifulSoup
import traceback
import pandas as pd
import re

```

```python
def get_list(text,year):
    soup=BeautifulSoup(text,'lxml')
    main_text = soup.find_all('td',class_='content')[0].text.strip().split('\n') #正文
    big_news=[]
    for one_line in main_text:
        try:
            date=re.findall('\d+月\d+日',one_line)[0]
        except:
            date=0
        big_news+=[[year,date,one_line]]
    return big_news

tot_list=[]
for file in os.listdir('./大事记htmls/'):
    if file[-4:]!='html':
        continue
    year=file[:4]
    with open('./大事记htmls/'+file,'r',encoding='utf-8') as f:
        text=f.read()
        f.close()
    res=get_list(text,year)
#     print(res)
    tot_list+=res

res_df=pd.DataFrame(tot_list,columns=['year','date','title'])
year
res_df
```

```python
new_date=res_df['year']+res_df['date'].apply(lambda x:'-'+'-'.join(re.findall('\d+',str(x))))
res_df['new_date']=new_date
```

```python
res_df.to_excel('大事记.xlsx')
```
